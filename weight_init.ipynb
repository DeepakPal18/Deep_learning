```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
from sklearn.datasets import make_moons
x, y = make_moons(n_samples=500,noise=0.05,random_state=101)
```

```python
# sns.scatterplot(x[:,0],x[:,1],hue=y)
plt.scatter(x[:,0],x[:,1],c = y)
plt.show()
```

```python
# API - 
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(2,activation='sigmoid',input_dim=2)) # don't give sigmoid , give  'relu'.
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.get_weights()
```

```python
initial_weights = model.get_weights()
```

```python
# taking constant value for all weights -
initial_weights[0] = np.ones(initial_weights[0].shape)*0.5
initial_weights[1] = np.ones(initial_weights[1].shape)*0.5
initial_weights[2] = np.ones(initial_weights[2].shape)*0.5
initial_weights[3] = np.ones(initial_weights[3].shape)*0.5
```

```python
model.set_weights(initial_weights)
```

```python
model.get_weights()
```

```python
x.shape,y.shape
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# spliting in train test with validation_split - 
history=model.fit(x,y,epochs=100,validation_split=0.2)
```

```python
model.get_weights()
```

```python
# getting wrong linear line - 
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y.astype('int'),clf=model,legend=2)
```

```python
# taking zero value for all weights -
initial_weights[0] = np.zeros(initial_weights[0].shape)*0.5
initial_weights[1] = np.zeros(initial_weights[1].shape)*0.5
initial_weights[2] = np.zeros(initial_weights[2].shape)*0.5
initial_weights[3] = np.zeros(initial_weights[3].shape)*0.5
```

```python
model.set_weights(initial_weights)
```

```python
model.get_weights()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x,y,epochs=100,validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y.astype('int'),clf=model,legend=2)
```

#### Proper weight initialization - (he-init)



```python
# using he_init for normal distribution - 
model = Sequential()
model.add(Dense(2,activation='relu',input_dim=2,kernel_initializer='he_normal')) 
model.add(Dense(2,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
# using he_init for uniform distribution - 
model = Sequential()
model.add(Dense(2,activation='relu',input_dim=2,kernel_initializer='he_uniform')) 
model.add(Dense(2,activation='relu',kernel_initializer='he_uniform'))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.get_weights()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x,y,epochs=100,validation_split=0.2)
```

```python
# getting more accurate decision line -
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y.astype('int'),clf=model,legend=2)
```
