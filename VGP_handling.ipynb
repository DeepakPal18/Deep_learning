```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

```python
# inbuild datasets 
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
# API sequential
from keras.layers import Dense
from keras.models import Sequential
```

```python
x, y = make_moons(n_samples = 250, noise=0.05, random_state=101)
```

```python
x
```

```python
y
```

```python
plt.scatter(x[:,0],x[:,1], c=y,s=100)
plt.show()
```

```python
# complex model with many hidden function - 
model = Sequential()
model.add(Dense(10,activation = 'sigmoid',input_dim=2))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(10,activation = 'sigmoid'))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.get_weights()[0] # 1st hidden layer
```

```python
old_weights = model.get_weights()[0]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2,random_state=101)
```

```python
#vanashing gradient problem loss not changing 
model.fit(x_train,y_train,epochs =500) 
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = (y_pred >0.5)
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)
```

### Handling VGP - 

#### 1) Reduce complexity(hidden layers)

```python
model1 = Sequential()
model1.add(Dense(10,activation='sigmoid',input_dim=2))
model1.add(Dense(10,activation='sigmoid'))
model1.add(Dense(10,activation='sigmoid'))
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model1.get_weights()[0]
```

```python
new_weights = model1.get_weights()[0]
```

```python
old_weights
```

```python
new_weights
```

```python
# reducing complexity(hidden layers) loss is reduced and accuracy has incresed
model1.fit(x_train,y_train,epochs=500)
```

```python
y_pred = model1.predict(x_test)
```

```python
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

#### 2) Changing activation function to relu

```python
# complex model with many hidden function - but changing activation as relu
model2 = Sequential()
model2.add(Dense(10,activation = 'relu',input_dim=2))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(10,activation = 'relu'))
model2.add(Dense(1,activation = 'sigmoid')) # as output is binary so sigmoid
```

```python
model2.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model2.get_weights()[0]
```

```python
relu_weights = model2.get_weights()[0]
```

```python
# relu improves accuracy and decreases loss value , incase relu doesn't give 
# good result use leakyrelu as relu performs bad with negative values as
# it converts them to 0 . thus results in vgp
model2.fit(x_train,y_train,epochs=500)
```

```python
y_pred = model2.predict(x_test)
```

```python
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
